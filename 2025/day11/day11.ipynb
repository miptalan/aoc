```python
result2:0!select sum c by s,e from raze f2[routing2]\[routing2]

svr2dac:0^first exec c from result2 where s=`svr,e=`dac
svr2fft:0^first exec c from result2 where s=`svr,e=`fft
/ only 1 should be non-0 - b/c no cycles
fft2dac:0^first exec c from result2 where s=`fft,e=`dac
dac2fft:0^first exec c from result2 where s=`dac,e=`fft

dac2out:0^first exec c from result2 where s=`dac,e=`out
fft2out:0^first exec c from result2 where s=`fft,e=`out
show `long$(svr2dac;dac2fft;fft2out;svr2fft;fft2dac;dac2out);
show ""
show `long$(svr2dac*dac2fft*fft2out)+svr2fft*fft2dac*dac2out;
```
