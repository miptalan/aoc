# Part 1
Rely on even-digitness and divisibility of 12341234 by 10001

```python
/ only even-number-digit numbers are of interest
{[(mn;mx)] s where 0=mod[ceiling 10 xlog s:mn+til 1+mx-mn;2]} 0N!input 2
```

```python
/ symmetrical number abab (3rd power of 10) should be divisible by 101: 1+10^(1+3)/2
{[(mn;mx)] s where (0=mod[s;1+10 xexp n2 div 2])&0=mod[n2:ceiling 10 xlog s:mn+til 1+mx-mn;2]} input 2
```

```python
/ solution
sum raze {[(mn;mx)] s where (0=mod[s;1+10 xexp n2 div 2])&0=mod[n2:ceiling 10 xlog s:mn+til 1+mx-mn;2]} each input
```

Now, an ID is invalid if it is made only of some sequence of digits repeated at least twice. So, 12341234 (1234 two times), 123123123 (123 three times), 1212121212 (12 five times), and 1111111 (1 seven times) are all invalid IDs.

From the same example as before:

*   11-22 still has two invalid IDs, 11 and 22.

*   222220-222224 still has one invalid ID, 222222.
    
*   1698522-1698528 still contains no invalid IDs.
    
*   446443-446449 still has one invalid ID, 446446.
    
*   38593856-38593862 still has one invalid ID, 38593859.
    
*   565653-565659 now has one invalid ID, 565656.
    
*   824824821-824824827 now has one invalid ID, 824824824.
    
*   2121212118-2121212124 now has one invalid ID, 2121212121.
    

Adding up all the invalid IDs in this example produces 4174379265.

What do you get if you add up all of the invalid IDs using these new rules?

2 digits - only 11  -> aa

3 digits - only 111 -> aaa

4 digits - 101 and 1111 -> abab and aaaa

5 digits - only 11111 -> aaaaa

6 digits - 1001, 10101 and 111111 -> abcabc, ababab and aaaaaa

*in general it is pair-wise factorisations*

```python
/example for 12
111111111111 div 1 / 12 1-digits
121212121212 div 12 / 6 2-digits
123123123123 div 123 / 4 3-digits
123412341234 div 1234 / 3 4-digits
123456123456 div 123456 / 2 6-digits
```

```python
/ simple factorisation
ffn:{-1_flip[(x div d;d)]idx:where 0=x mod/:d:1+til x}
show factors:ffn 12
```

```python
/ generate "Nth power of 10 shited m times" pattern
{[d;p] 1+{y+x*y}/[p10;(d-2)#p10:prd p#10]}[4;3]
```

```python
/ example for 12, 10, 3 and 2
{[d;p] 1+{y+x*y}/[p10;(d-2)#p10:prd p#10]}.'ffn 12
{[d;p] 1+{y+x*y}/[p10;(d-2)#p10:prd p#10]}.'ffn 10
{[d;p] 1+{y+x*y}/[p10;(d-2)#p10:prd p#10]}.'ffn 3
{[d;p] 1+{y+x*y}/[p10;(d-2)#p10:prd p#10]}.'ffn 2
```

```python
/ pre-generate divisors
k:1_1+til max ceiling 10 xlog input[;1]
d:k!{{[d;p] 1+{y+x*y}/[p10;(d-2)#p10:prd p#10]}.'ffn x}each k
d
```

```python
/ getting divisors 
{[(mn;mx)] d ceiling 10 xlog s:0N!mn+til 1+mx-mn} input 2
```

```python
/ any divisor getting 0 remainder?
raze {[(mn;mx)] s where any each 0=s mod'd ceiling 10 xlog s:mn+til 1+mx-mn} input 2
```
