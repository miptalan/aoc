```python
import numpy as np
fname = "input.txt"
with open(fname) as f:
    algo = np.array([1 if x=="#" else 0 for x in f.readline().strip()])
    f.readline()
    img = np.array([[1 if x=="#" else 0 for x in y.strip()] for y in f.readlines()])
    
def v(img):
    _ = [print("".join(f)) for f in np.where(img, "#", ".")]
    
algo
```

```python
v(img)
```

```python
def get3x3(a, i, j):
    res = []
    for ii in range(i-1,i+2):
        for jj in range(j-1,j+2):
            if ii<0 or ii>=len(a) or jj<0 or jj>=len(a):
                res.append(a[0,0])
            else:
                res.append(a[ii,jj])
    b = np.packbits(res)
    return np.left_shift(b[0], 1) + np.right_shift(b[1], 7)
```

```python
def enhance(inp, f=0, algo=algo):
    inp = np.pad(inp, 1, mode='constant', constant_values=f)
    outp = np.zeros_like(inp)
    for i in range(0, len(inp)):
        for j in range(0, len(inp)):
            outp[i,j] = algo[get3x3(inp, i, j)]

    return outp
```

```python
v(enhance(img))
```

```python
v(enhance(enhance(img)))
```

```python
def nhance(img, n, algo=algo):
    f = 0
    for i in range(n):
        img = enhance(img, f)
        if algo[0]:
            f = algo[0] if i%2==0 else algo[-1]
    return np.sum(img)
```

```python
%timeit nhance(img, 2)
```
